## Visualization & Explanatory Analysis

### Pertanyaan 1:

1. jam yang minat sewanya tinggi

```python
sns.set_style("whitegrid")
plt.figure(figsize=(11, 5))
sns.lineplot(x="hr", y="cnt", data=df_hour, ci=None)
plt.title("Jam sewa")
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.show()
```

### Pertanyaan 2:

1. Apakah ada Korelasi antara cuaca dan jumlah sewa sepeda?

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="weathersit", y="cnt", hue="weathersit", data=df_day, palette="Set3")
plt.title("Korelasi cuaca dengan jumlah sewa sepeda")
plt.xlabel("Weathersit")
plt.ylabel("Jumlah Sewa Sepeda")
plt.show()
```

## Conclusion
